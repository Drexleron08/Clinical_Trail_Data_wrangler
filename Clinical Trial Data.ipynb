```python
#creating synthetic datasets for a clinical trial on a new drug for hypertension
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)
n_patients = 550

# 1. Demographics
demo = pd.DataFrame({
    'Patient_ID': range(101, 101 + n_patients),
    'Age': np.random.randint(15, 85, n_patients),
    'Sex': np.random.choice(['M', 'F'], n_patients),
    'Baseline_BP': np.random.randint(110, 160, n_patients)
})
# Manually add some "messy" data (NaNs)
demo.loc[120:124, 'Age'] = np.nan 
demo.loc[116:124, 'Baseline_BP'] = np.nan 


# 2. Lab Results
labs = pd.DataFrame({
    'Patient_ID': range(101, 101 + n_patients),
    'Cholesterol': np.random.randint(150, 300, n_patients),
    'Glucose': np.random.randint(70, 200, n_patients)
})
#manually add some "messy" data (NaNs)
labs.loc[10:15, 'Glucose'] = np.nan

# 3. Dosage Logs
dosage = pd.DataFrame({
    'Patient_ID': range(101, 651),
    'Drug_Type': np.random.choice(['Placebo', 'Drug_A'], 550),
    'Dose_mg': [5, 10, 20, 10, 5] * 110
})

# Save to CSVs
demo.to_csv('demographics.csv', index=False)
labs.to_csv('lab_results.csv', index=False)
dosage.to_csv('dosage_logs.csv', index=False)
```
